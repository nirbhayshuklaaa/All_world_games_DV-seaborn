```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Top 10 Countries Vs Participations 

df=pd.read_csv("Game.csv")
data=df["Team"].value_counts().head(10)
plt.figure(figsize=(16,7))
sns.barplot(hue=data.index,x=data.values,orient="h",edgecolor='black')
plt.xlabel("No. of Participations")
plt.ylabel("Teams")
plt.title("Top 10 Countries Vs Participations")
plt.savefig("Top 10 Countries Vs Participations.png",dpi=300)
plt.show()
```

```python
# Top 10 Played game in dataset 

sport=df['Sport'].value_counts().head(10)
plt.figure(figsize=(16,8))
sns.barplot(hue=sport.index,y=sport.values,palette='Set3',saturation=1,edgecolor='black')
plt.ylabel("Count")
plt.xlabel("Sports")
plt.title("Top 10 Played game")
plt.savefig("top 10 Played Games.png",dpi=300)
plt.show()
```

```python
# Season Vs Games played 

season=df['Season'].value_counts()
plt.pie(season,autopct="%0.1f%%",labels=season.index,colors=['y','c'],shadow=True)
plt.title("Seasons Vs Games Played")
plt.savefig("Season vs games Ration.png",dpi=300)
plt.show()
```

```python
# Gender participation Ratio
gender=df['Sex'].value_counts()
gender
plt.pie(gender,labels=gender.index,autopct="%0.1f%%",colors=['pink','red'],shadow=True)
plt.title("Gender Participant ratio")
plt.savefig("Gender participation.png",dpi=300)
plt.show()
```

```python
# Get top 5 categories

top5 = df['Sport'].value_counts().head(5)
plt.figure(figsize=(8,8))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%', startangle=90)

plt.title("Top 5 Sports Distribution")
plt.savefig("Top 5 categories Ratio.png",dpi=300)
plt.show()
```

```python
#Gender vise played Game

sns.countplot(x='Sport', hue='Sex', data=df,
              order=df['Sport'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title("Gender vise played Game")
plt.savefig("Gender vise playings.png",dpi=300)
plt.show()
```

```python
# age Disrtibution of Players 

sns.histplot(df['Age'],bins=20,kde=True,color='c')
plt.title ("Age Distribution of Players")
plt.savefig("Age distribution.png",dpi=300)
plt.show()
```

```python
# Age wise medal Won by gender

sns.scatterplot(x='Age',y='Medal',data=df,hue='Sex')
plt.title("Age wise medal Won by gender")
plt.savefig("Age vise medals.png",dpi=300)
plt.show()
```

```python
# Age wise male femmale played in top 10 games
plt.figure(figsize=(16,10))
sns.violinplot(x='Sport', y='Age', order=df['Sport'].value_counts().index[0:10], data=df,hue='Sex',palette="spring",saturation=2)
plt.title("Age wise male femmale played in top 10 games")
plt.savefig("Age wise male femmale played.png",dpi=300)
plt.show()
```
